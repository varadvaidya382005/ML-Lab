# For feeling null values

```python

threshold = 0.5 * len(dataset)
dataset_1 = dataset.dropna(thresh=threshold, axis=1)


dataset_1.fillna(dataset_1.mean(numeric_only=True), inplace=True)
```

# Dropping columns which are not required

```python
columns_to_drop = ["TransactionID", "CardSecurityCode", "PrimaryEmailDomain"]
dataset_1.drop(columns=[col for col in columns_to_drop if col in dataset_1.columns], inplace=True)
```

```python
dataset_1.isnull().sum()
```

# Label Encoding

```python

dataset_1 = dataset_1.copy()
if "CardType" in dataset_1.columns:
    dataset_1["CardType"] = dataset_1["CardType"].fillna("Unknown")

if "CardUsageType" in dataset_1.columns:
    dataset_1["CardUsageType"] = dataset_1["CardUsageType"].fillna("Unknown")

if "CardSecurityCode" in dataset_1.columns:
    dataset_1["CardSecurityCode"] = dataset_1["CardSecurityCode"].fillna(0)  # Assuming numeric
```

```python
dataset_1.isnull().sum()
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = dataset_1.select_dtypes(include=['number']).columns
dataset_1[num_cols] = scaler.fit_transform(dataset_1[num_cols])
```

# Label Encoding the columns which has categorical values

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()


for col in dataset_1.select_dtypes(include=['object']).columns:
    dataset_1[col] = label_encoder.fit_transform(dataset_1[col])
```
